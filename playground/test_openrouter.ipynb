```python
import requests
import json

response = requests.post(
    'https://openrouter.ai/api/v1/responses',
    headers={
        'Authorization': 'Bearer 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',
        'Content-Type': 'application/json',
    },
    json={
        'model': 'openai/gpt-oss-20b:free',
        'input': 'Solve this step by step: If a train travels 60 mph for 2.5 hours, how far does it go?',
        'reasoning': {
            'effort': 'high'
        },
        'stream': True,
        'max_output_tokens': 9000,
    },
    stream=True
)

# 显式打印返回的原始内容用于调试
has_output = False
for line in response.iter_lines():
    if line:
        has_output = True
        try:
            line_str = line.decode('utf-8')
            # print(f"RAW LINE: {line_str}")  # 打印每行内容，方便调试
            if line_str.startswith('data: '):
                data = line_str[6:]
                if data == '[DONE]':
                    print("流式响应已结束。")
                    break
                try:
                    parsed = json.loads(data)
                    print(f"解析内容: {parsed}")  # 打印解析出的json
                    if parsed.get('type') == 'response.reasoning.delta':
                        print(f"Reasoning: {parsed.get('delta', '')}")
                    elif parsed.get('type') == 'response.delta':
                        print(f"模型回复: {parsed.get('delta', '')}")
                except json.JSONDecodeError as jde:
                    print(f"[JSON解析错误]: {jde} -- 数据内容: {data}")
        except Exception as e:
            print(f"[行处理异常]: {e}")

if not has_output:
    print("警告：服务器无流式内容输出，检查网络、API KEY、配额和参数设置。")
```
